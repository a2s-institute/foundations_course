# Anatomy of a Plot

The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures and Figures can contain multiple Axes.

Most plotting ocurs on an Axes. The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

Each Axes has an XAxis and a YAxis. These contain the ticks, tick locations, labels, etc.

# Axes

All plotting is done with respect to an Axes. An Axes is made up of Axis objects and many other things. An Axes object must belong to a Figure (and only one Figure). Most commands you will ever issue will be with respect to this Axes object.

Typically, you'll set up a Figure, and then add an Axes to it.

```python
fig = plt.figure()
ax = fig.add_subplot(111) # Basically, 1 row and 1 column.
ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], title='Axes/Subplots',
       ylabel='Y-Axis', xlabel='X-Axis')
```

# Basic Plotting

Most plotting happens on an Axes. Therefore, if you're plotting something on an axes, then you'll use one of its methods. 
Plot draws points with lines connecting them.  scatter draws unconnected points, optionally scaled or colored by additional variables.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(title="Sample Plot", xlabel="X axis", ylabel="Y axis",
      xlim=[0.5, 4.5], ylim=[-2, 8],
      xticks=np.arange(0.5, 5, 0.5),
      yticks=np.arange(-2, 9, 1))
ax.plot([1, 2, 3, 4], [1, 2, 5, 8], color='lightblue', linewidth=2, label="line plot")
ax.scatter([0.6, 3.8, 1.2, 2.5], [1, 5, 7, 6], color='darkgreen', marker='o', label="scatter plot")
ax.legend(loc='lower right')
ax.grid()
```

# Multiple Axes

A figure can have more than one Axes on it.

```python
x = np.linspace(-1.4, 1.4, 30)

fig = plt.figure()

ax = fig.add_subplot(221) # 2 rows, 2 columns, 1st subplot = top left
ax.plot(x, x)

ax = fig.add_subplot(222) # 2 rows, 2 columns, 2nd subplot = top right
ax.plot(x, x ** 2)

ax = fig.add_subplot(223) # 2 rows, 2 columns, 3rd subplot = bottow left
ax.plot(x, x ** 3, color='lightblue', linewidth=2)

ax = fig.add_subplot(224) # 2 rows, 2 columns, 4th subplot = bottom right
ax.plot(x, x ** 4, label="asdf")



ax.legend()

plt.tight_layout() # Make the subplots fill up the figure a bit more
```

```python
fig = plt.figure(figsize=(10,10))
x = np.linspace(-180,180)


axsin = fig.add_subplot(334)
axsin.set(title="sin(x)", xlabel="x in degrees", ylabel="y=sin(x)",
      xlim=[-180, 180], ylim=[-1, 1])
axsin.plot(x, np.sin(np.deg2rad(x)))

axcos = fig.add_subplot(336)
axcos.set(title="cos(x)", xlabel="x in degrees", ylabel="y=cos(x)",
      xlim=[-180, 180], ylim=[-1, 1])
axcos.plot(x, np.cos(np.deg2rad(x)))
```

# Non Linear Scales

```python
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(121)
ax.plot(x, y)
ax.set_yscale('linear')
ax.set_title('linear')
plt.grid(True)

ax = fig.add_subplot(122)
ax.plot(x, y)
ax.set_yscale('log')
ax.set_title('log')
plt.grid(True)

plt.tight_layout()
```

# 3D Projection

Plotting 3D graphs is quite straightforward. You need to import Axes3D, which registers the "3d" projection. Then create a subplot setting the projection to "3d". This returns an Axes3DSubplot object, which you can use to call plot_surface, giving x, y, and z coordinates, plus optional attributes.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
# X, Y = np.meshgrid(x, y)
X,Y = np.broadcast_arrays(x,y.reshape(y.shape[0],1))
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize=(12, 4))
subplot3d = plt.subplot(111, projection='3d')

surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0.1)
```

### Intro to Sympy

Sympy is a python library for symbolic mathematics, what this means is that we can solve complex calculus, linear algebra equations etc. using their exact mathematical representations.  

```python
import sympy as sp
sp.init_printing(use_latex=True)           # for latex printing 
```

Variables in sympy must be defined as follows before being used in an equation.

```python
# Here we initalize the symbols that shall be used in the equation
x,y = sp.symbols('x y')
```

```python
# We can now form the equation using the variables we declared previously
eqn = x + 2*y

# We can now view the stored equation as 
print ("Initial equation :",eqn)

# We can also play around with the equation by adding or removing terms 
print ("Equation with a new term:",eqn + 45)

print ("Equation with the x term removed:",eqn-x)
```

To obtain the derivative of a function, in sympy we can simply use the inbuilt differentiation function. A similar inbuilt function can also be used to calculate the integral of a given equation.

```python
# Here we obtain the derivative of the sine function
sp.diff(sp.sin(x),x)
```

We can also solve ordinary differential equations such as:  $y'-y = e^t $ in the following manner

```python
# First we declare the spmbols required for the equation
t = sp.symbols('t')

# Here we declare the symbolic function ===> y (dependent variable) is a function of t (independent variable)
y = sp.Function('y')(t)

y_prime = y.diff(t, 1)            # Finding the first derivative. Use 2 for second derivative and 3 for third derivative and so on...

# Now we declare the equation to be solved
eq = y_prime - y - sp.exp(t)

# We call the dsolve method that solves the equation with respect to the function y(t)
sp.dsolve(eq)
```

Similarly to solve an algebraic equation such as: $$ {x^2} = 1  $$ we can use the inbuilt solve function once we have sucessfully declared the variables and the equation itself.

```python
# Declare the equation using the symbols we declared earlier
eqn = x**2 - 1

# Calling the solve function on the declared equation
sp.solve(eqn)
```

In order to evaluate an equation using a given set of values we use the evalf method.

```python
# Lets evaluate the equation we declared in the previous cell
eqn.evalf(subs = {x:5})
```
