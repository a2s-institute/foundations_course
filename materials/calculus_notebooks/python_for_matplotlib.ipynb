# Anatomy of a Plot

The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures and Figures can contain multiple Axes.

Most plotting ocurs on an Axes. The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

Each Axes has an Xaxis and a Yaxis. These contain the ticks, tick locations, labels, etc. 

Further reading : https://matplotlib.org/stable/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py

### Here we have intiated a figure with a single Axes

```python
fig, ax = plt.subplots()  # a figure with a single Axes
```

You can control the size of the figure through the figsize argument, which expects a tuple of (width, height) in inches.

```python
fig, ax = plt.subplots(figsize = (9, 5))  #here te size of the figure is set by the figsize argument.
```

# Axes

All plotting is done with respect to an Axes. An Axes is made up of Axis objects and many other things. An Axes object must belong to a Figure (and only one Figure). Most commands you will ever issue will be with respect to this Axes object. With this Axes object, we can add title, labes for different axis, legends etc.

In the above command, ax is the Axes object.

# Basic Plotting

Most plotting happens on an Axes. Therefore, if you're plotting something on an axes, then you'll use one of its methods. 
Plot draws points with lines connecting them.  scatter draws unconnected points, optionally scaled or colored by additional variables.

```python
#making some data to plot
x = [1 ,2 ,3 ,4, 5]

# plotting data
fig, ax = plt.subplots(figsize = (9, 5))  # Create a figure and an axes.
ax.plot(x, x)  # Plot some data on the axes.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
plt.show()
```

# Multiple Axes

A figure can have more than one Axes on it.

```python
#generating synthetic data
x = [1 ,2 ,3 ,4, 5]
x_2 = [1, 4, 9, 16, 25]

#creating figure with 2 axes. here the 2 axes are divided into 1 row and 2 columns
fig, axs = plt.subplots(2,2,figsize = (15, 5)) 

#plotting data on the first axes
axs[0,0].plot(x, x) 
axs[0,0].set_xlabel('x label')  # Add an x-label to the axes.
axs[0,0].set_ylabel('y label')  # Add a y-label to the axes.
axs[0,0].set_title("Simple Plot")  # Add a title to the axes.

#plotting data on the Second axes
axs[0,1].plot(x, x_2)  
axs[0,1].set_xlabel('x label  2')  # Add an x-label to the axes.
axs[0,1].set_ylabel('y label  2')  # Add a y-label to the axes.
axs[0,1].set_title("Simple Plot 2")  # Add a title to the axes.

#plotting data on the first axes
axs[1,0].plot(x, x) 
axs[1,0].set_xlabel('x label')  # Add an x-label to the axes.
axs[1,0].set_ylabel('y label')  # Add a y-label to the axes.
axs[1,0].set_title("Simple Plot")  # Add a title to the axes.

#plotting data on the Second axes
axs[1,1].plot(x, x_2)  
axs[1,1].set_xlabel('x label  4')  # Add an x-label to the axes.
axs[1,1].set_ylabel('y label  4')  # Add a y-label to the axes.
axs[1,1].set_title("Simple Plot 4")  # Add a title to the axes.
plt.show()
```

# Legends

The simplest way to add a legend is to set a label on all lines, then just call the legend function.

```python
#making some data to plot
x = [1 ,2 ,3 ,4, 5]
x_2 = [1, 4, 9, 16, 25]
x_3 = [1, 8, 27, 64, 125]

# plotting data

fig, ax = plt.subplots(figsize = (9, 5))  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x_2, color = 'purple', label='quadratic')  # Plot more data on the axes with purple color
ax.scatter(x, x_3, color = 'orange',marker = '*', label='cubic')  # and some more with scatter, orange color, and different marker
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Calling the legend function
plt.show() #showing the plot.
```

# Non Linear Scales

```python
#making some data to plot
x = [1 ,2 ,3 ,4, 5]
y = [1, 10, 100, 1000, 10000]
# plotting data
fig, ax = plt.subplots(figsize = (9, 5))  # Create a figure and an axes.
ax.plot(x, x)  # Plot some data on the axes.
ax.set_xlabel('log x')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_xscale('log') #setting the x axis to log
ax.set_title("Plot with x-axis in Log")  # Add a title to the axes.
plt.show()



# plotting data
fig, ax = plt.subplots(figsize = (9, 5))  # Create a figure and an axes.
ax.plot(x, y)  # Plot some data on the axes.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('log y')  # Add a y-label to the axes.
ax.set_yscale('log')
ax.set_title("Plot with x-axis in Log")  # Add a title to the axes.
plt.show()
```

# 3D Projection

Plotting 3D graphs is quite straightforward. 3D plots are enabled by importng the mplot3d toolkit. Once this module have been imported then we can create a 3D axes by passing projection='3d' in any axes creation routine.

```python
# https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html
from mpl_toolkits import mplot3d
import numpy as np
#creating figure
fig = plt.figure(figsize = (10, 6))
#creating 3d axes
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
